## Introduction/Business Problem

Buenos Aires is the biggest city of Argentina. The entire urban conglomerate has 17 million people, the Capital District is its main part and has 3 milloon people.

The Capital is a tourist pole with 5.5 millio visitors every year. Buenos Aires has a wide variety of attractions, services and places. It has 50 neighborhoods, everyone is different, but Â¿can we group them by similarity?
Of course, yes. How homogeneous is the city? Where can we find those neighborhoods especially different from the rest?

We want to detect the differences between some naighborhood clusters to help local business to exploit the cultural advantages of your neighborhood, and suggest to tourists where to look for new experiences in the city.

## Data

We will use a CSV file that contains everyone of the 50 Buenos Aires' neighbourhoods and their respective Latitude and Lingitude. Then we will combine it with Foursquare information to be able of compare the different neighbourhoods.
We are interested in the venues more common in every neighborhood. These will allow us to create 5 clusters.

## Methodology

Our objective is to compare the 50 Buenos Aires' neighbourhoods and establish 5 different clusters that can describe the city based on its venues.

To reach this objective we will use both Kmeans clustering and Hierarchical clustering. In order to do that, we have to pre-process the data and combine our original "barrios" dataset with the venue information provided by Foursquare. We will display the Buenos Aires' map with the neighborhood clusters.

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues)
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
# lets see some Palermo's venues
nearby_venues.head()
```

### create a sistem to search venues in every neighborhood

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Barrio', 
                  'Barrio Latitud', 
                  'Barrio Longitud', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
ba_venues = getNearbyVenues(names=barrios['Barrio'],
                                   latitudes=barrios['Latitud'],
                                   longitudes=barrios['Longitud']
                                  )
```

```python
# lets see some random venues
ba_venues.sample(10)
```

```python
# one hot encoding
ba_onehot = pd.get_dummies(ba_venues[['Venue Category']], prefix="", prefix_sep="")

# add borough column back to dataframe
ba_onehot['Barrio'] = ba_venues['Barrio'] 

# move borough column to the first column
fixed_columns = [ba_onehot.columns[-1]] + list(ba_onehot.columns[:-1])
ba_onehot = ba_onehot[fixed_columns]

ba_grouped = ba_onehot.groupby('Barrio').mean().reset_index()
ba_grouped.sample(10)
```

### lets take 5 random neighbours, and see they most common venues

```python
for i in ba_grouped['Barrio'].sample(5):
    print("----"+i+"----")
    temp = ba_grouped[ba_grouped['Barrio'] == i].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(3))
    print('\n')
```

### claster the neughbourhoods with Kmeans

```python
kclusters = 5
ba_cluster = ba_grouped.drop('Barrio', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=135).fit(ba_cluster)

kmeans.labels_[::]
```

```python
barrios=pd.read_csv("buenosaires.csv")
barrios.insert(1, 'Cluster Kmeans', kmeans.labels_)
```

```python
barrios.sample(5)
```

```python
map_clusters = folium.Map(location=[buenos_aires.latitude, buenos_aires.longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(barrios['Latitud'], barrios['Longitud'], barrios['Barrio'], barrios['Cluster Kmeans']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

## Discussion

We observe that, as we see, every neighborhood has quite a variety of different venues. The city in general is pretty homogeneous: East, center and nort zones has the same cluster (cluster 4 in K means). Both K-means and Hierarchical Clustering shows this phenomenon.

Nor-west, west and south neighborhoods where classified in other clusters. That neighborhoods are essentially different from the central and coastal neighborhoods. Hierarchical clustering shows  that more robustly, especially in the south-eastern areas of the city.

## Conclusions

The city of Buenos Aires is a well-equipped area in terms of entertainment, recreation and gastronomy.
The neighborhoods of the downtown and coastal area are similar to each other. The more remote areas show greater variety, in terms of venues.
